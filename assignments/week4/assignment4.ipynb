```python
import pandas as pd
import numpy as np
import scipy.stats as stats
import re

mlb_df=pd.read_csv("assets/mlb.csv")
nhl_df=pd.read_csv("assets/nhl.csv")
nba_df=pd.read_csv("assets/nba.csv")
nfl_df=pd.read_csv("assets/nfl.csv")
cities=pd.read_html("assets/wikipedia_data.html")[1]
cities=cities.iloc[:-1,[0,3,5,6,7,8]]

## NHL 
def nhl_correlation(): 
    # YOUR CODE HERE
    nhl_df=pd.read_csv("assets/nhl.csv")
    cities=pd.read_html("assets/wikipedia_data.html")[1]
    cities=cities.iloc[:-1,[0,3,5,6,7,8]]
    
    cities['Metropolitan area'] = cities['Metropolitan area'].str.replace(r' $', '', regex=True)
    # cities['NHL'] = cities['NHL'].str.replace(r'\—|\[|\]|', '', regex = True)
    # cities.NHL.unique()
    
    nhl_df_filtered = nhl_df[(nhl_df['year'] == 2018)]
## Removed rows with Division
    nhl_df_filtered = nhl_df_filtered[~(nhl_df_filtered ['team'].str.contains('Division'))]

#     print (sorted(cities['Metropolitan area'].unique()))
#     print (len(cities['Metropolitan area'].unique()))
    
    nhl_df_filtered['team'] = nhl_df_filtered['team'].str.replace(r'\*', '', regex=True)

    # Convert the NHL Team column to map to the State
    nhl_df_filtered['team_ville'] = nhl_df_filtered['team']
#     print (len(nhl_df_filtered['team'].unique()))
    nhl_df_filtered['team_ville'] = nhl_df_filtered['team_ville'].map({'Tampa Bay Lightning':'Tampa Bay Area',
         'Boston Bruins':'Boston',
         'Toronto Maple Leafs':'Toronto',
         'Florida Panthers':'Miami–Fort Lauderdale',
         'Detroit Red Wings':'Detroit',
         'Montreal Canadiens':'Montreal',
         'Ottawa Senators':'Ottawa',
         'Buffalo Sabres':'Buffalo',
         'Washington Capitals':'Washington, D.C.',
         'Pittsburgh Penguins':'Pittsburgh',
         'Philadelphia Flyers':'Philadelphia',
         'Columbus Blue Jackets':'Columbus',
         'New Jersey Devils':'New York City',
         'Carolina Hurricanes':'Raleigh',
         'New York Islanders':'New York City',
         'New York Rangers':'New York City',
         'Nashville Predators':'Nashville',
         'Winnipeg Jets':'Winnipeg',
         'Minnesota Wild':'Minneapolis–Saint Paul',
         'Colorado Avalanche':'Denver',
         'St. Louis Blues':'St. Louis',
         'Dallas Stars':'Dallas–Fort Worth',
         'Chicago Blackhawks':'Chicago',
         'Vegas Golden Knights':'Las Vegas',
         'Anaheim Ducks':'Los Angeles',
         'San Jose Sharks':'San Francisco Bay Area',
         'Los Angeles Kings':'Los Angeles',
         'Calgary Flames':'Calgary',
         'Edmonton Oilers':'Edmonton',
         'Vancouver Canucks':'Vancouver',
         'Arizona Coyotes':'Phoenix'})
    df = pd.merge(nhl_df_filtered, cities, how = 'inner', left_on = 'team_ville', right_on='Metropolitan area')
    
    # Win loss ratio 
    df['W'] = df['W'].astype(int)
    df['L'] = df['L'].astype(int)
    df['Population (2016 est.)[8]'] = df['Population (2016 est.)[8]'].astype(int)
    ## Subset columns
    he = ['team','W','L','Metropolitan area','Population (2016 est.)[8]']
    df = df[he]
    df['W/L Ratio'] = df['W']/(df['L']+df['W'])
    # Average the winloss for the cities which have multiple teams
    df = df.groupby('Metropolitan area').mean().reset_index()
    return df

## NBA
def nba_correlation():
    # YOUR CODE HERE
    nba_df=pd.read_csv("assets/nba.csv")
    cities=pd.read_html("assets/wikipedia_data.html")[1]
    cities=cities.iloc[:-1,[0,3,5,6,7,8]]
    cities['Metropolitan area'] = cities['Metropolitan area'].str.lstrip()#.str.replace(r' $', '', regex=True)
    
    ## Get the last Team Name
    team_list = []
    for t in nba_df['team']:
        l =  re.findall(r"[A-Z][a-z]+(?= ).*|[A-Z][a-z]+(?! )", t)
        match_list = ''.join(l)
        # print(match_list)
        # print ("Before stripped")
        stripped = re.sub(r'[?=\\*]\s', '', match_list.split(r'[?=\\*]')[0])
        stripped2 = re.split(r'\(', stripped)[0]
        # print (stripped2)
        team_list.append(stripped2)
    # Get Team's last name
    ls = []   
    for l in team_list:
        split_str = l.split (' ',)
        # print (split_str)
        # if there are 3, take the last
        if len(split_str) < 3:
            last_name  = split_str[1].split('\\')[0].strip() # remove additional whitespaces
        elif len(split_str) == 3:
            last_name = split_str[2].split('\\')[0].strip() # remove additional whitespaces
        ls.append(last_name)
    # Convert to team last name
    nba_df['team_last_name'] = ls
    
    # Regex fill Cities name
    nba_cities = cities.copy()
    nba_cities = nba_cities[['Metropolitan area', 'NBA', 'Population (2016 est.)[8]']]
    nba_team = []
    for n in nba_cities['NBA']:
        ls = re.sub(r'\[[^()]*\]', '', n)
        split_strings = ''.join(re.split(' ', ls))
        nba_team.append(split_strings)
    nba_cities['NBA_team_dict'] = nba_team
    df = nba_cities
    
    # Expand into new rows
    out = df.assign(nba=df["NBA_team_dict"].str.findall(r"[A-Z](?:\s[A-Z]|[^A-Z])+")).explode("nba").reset_index(drop=True)
    out = out [['Metropolitan area', 'nba', 'Population (2016 est.)[8]']]
    # Remove philadelphia
    removed_phil = out[~(out['Metropolitan area'] =='Philadelphia')]
    # Fix to 76ers
    final_out = pd.concat([removed_phil, out[(out['Metropolitan area'] =='Philadelphia')].fillna('76ers')], ignore_index=True)
    final_out

    # Merge with the score nba
    nba_scores_metro = pd.merge(nba_df, final_out, left_on='team_last_name', right_on='nba')
    nba_scores_metro['W/L Ratio'] = nba_scores_metro['W'].astype(int) / (nba_scores_metro['L'].astype(int) + \
        nba_scores_metro['W'].astype(int))
    
    # Filter year to only 2018
    nba_scores_metro = nba_scores_metro[(nba_scores_metro['year'] == 2018)]
    nba_scores_metro_subset = nba_scores_metro [['year', 'W/L Ratio', 'Metropolitan area', 'Population (2016 est.)[8]']]
    nba_scores_metro_subset['Population (2016 est.)[8]'] = nba_scores_metro_subset ['Population (2016 est.)[8]'].astype(int)

    nba_scores_metro_subset = nba_scores_metro_subset.groupby('Metropolitan area').mean().reset_index()
    return nba_scores_metro_subset

## MLB 
def mlb_correlation(): 
    # YOUR CODE HERE
    mlb_df=pd.read_csv("assets/mlb.csv")
    cities=pd.read_html("assets/wikipedia_data.html")[1]
    cities=cities.iloc[:-1,[0,3,5,6,7,8]]

    ## Get the MLB last Team Name
    mlb_df.head()
    team_list = []
    for t in mlb_df['team']:
        l =  re.findall(r"[A-Z][a-z]+(?= ).*|[A-Z][a-z]+(?! )", t)
        match_list = ''.join(l)
        # print(match_list)
        # print ("Before stripped")
        stripped = re.sub(r'[?=\\*]\s', '', match_list.split(r'[?=\\*]')[0])
        stripped2 = re.split(r'\(', stripped)[0]
        # print (stripped2)
        team_list.append(stripped2)

    ls = []   
    for l in team_list:
        split_str = l.split (' ',)
        # print (split_str)
        # if there are 3, take the last
        if len(split_str) < 3:
            last_name  = split_str[1].split('\\')[0].strip()
        elif len(split_str) == 3:
            last_name = split_str[2].split('\\')[0].strip()
        ls.append(last_name)

    mlb_df['team_last_name'] = ls

    # Clean Cities for MLB
    mlb_cities = cities.copy()
    mlb_cities = mlb_cities[['Metropolitan area', 'MLB', 'Population (2016 est.)[8]']]
    ## Create list through regex, remove square brackts
    mlb_team = []
    for n in mlb_cities['MLB']:
        ls = re.sub(r'\[[^()]*\]', '', n)
        split_strings = ''.join(re.split(' ', ls))
        mlb_team.append(split_strings)
    mlb_cities['MLB_team_dict'] = mlb_team
    df = mlb_cities
    # Expand into new rows
    out = df.assign(mlb=df["MLB_team_dict"].str.findall(r"[A-Z](?:\s[A-Z]|[^A-Z])+")).explode("mlb").reset_index(drop=True)
    out = out [['Metropolitan area', 'mlb', 'Population (2016 est.)[8]']]

    # Merge with the score nba
    mlb_scores_metro = pd.merge(mlb_df, out, left_on='team_last_name', right_on='mlb')
    mlb_scores_metro['W/L Ratio'] = mlb_scores_metro['W'].astype(int) / (mlb_scores_metro['L'].astype(int) + \
        mlb_scores_metro['W'].astype(int))
    # Filter year to only 2018
    mlb_scores_metro = mlb_scores_metro[(mlb_scores_metro['year'] == 2018)]
    mlb_scores_metro_subset = mlb_scores_metro.copy()
    mlb_scores_metro_subset = mlb_scores_metro_subset [['year', 'W/L Ratio', 'Metropolitan area', 'Population (2016 est.)[8]']]
    mlb_scores_metro_subset['Population (2016 est.)[8]'] = mlb_scores_metro_subset ['Population (2016 est.)[8]'].astype(int)
    mlb_scores_metro_subset = mlb_scores_metro_subset.groupby('Metropolitan area').mean().reset_index()
    return mlb_scores_metro_subset

## NFL 
def nfl_correlation(): 
    # YOUR CODE HERE
    nfl_df=pd.read_csv("assets/nfl.csv")
    cities=pd.read_html("assets/wikipedia_data.html")[1]
    cities=cities.iloc[:-1,[0,3,5,6,7,8]]
    ## Get the last Team Name
    nfl_df.head()
    team_list = []

    ls = []   
    # for l in team_list:
    for l in nfl_df['team']:
        split_str = l.split (' ')
        last_name  = split_str[-1]#.split('')[0].strip()
        last_name = re.sub('\*|\+', '', last_name)
        ls.append(last_name)
    nfl_df['team_last_name'] = ls
    ## Fix metropolitan area for NFL
    nfl_cities = cities.copy()
    nfl_cities = nfl_cities[['Metropolitan area', 'NFL', 'Population (2016 est.)[8]']]

    nfl_team = []
    for n in nfl_cities['NFL']:
        ls = re.sub(r'\[[^()]*\]', '', n)
        split_strings = ''.join(re.split(' ', ls))
        nfl_team.append(split_strings)
    nfl_cities['NFL_team_dict'] = nfl_team
    df = nfl_cities


    # Expand into new rows
    out = df.assign(nfl=df["NFL_team_dict"].str.findall(r"[A-Z0-9](?:\s[A-Z]|[^A-Z])+")).explode("nfl").reset_index(drop=True)
    out = out [['Metropolitan area', 'nfl', 'Population (2016 est.)[8]']]

    # Merge with the score nfl
    nfl_scores_metro = pd.merge(nfl_df, out, left_on='team_last_name', right_on='nfl')
    nfl_scores_metro['W/L Ratio'] = nfl_scores_metro['W'].astype(int) / (nfl_scores_metro['L'].astype(int) + \
        nfl_scores_metro['W'].astype(int))
    # Filter year to only 2018
    nfl_scores_metro = nfl_scores_metro[(nfl_scores_metro['year'] == 2018)]
    nfl_scores_metro_subset = nfl_scores_metro.copy()
    nfl_scores_metro_subset = nfl_scores_metro_subset [['year', 'W/L Ratio', 'Metropolitan area', 'Population (2016 est.)[8]']]
    nfl_scores_metro_subset['Population (2016 est.)[8]'] = nfl_scores_metro_subset ['Population (2016 est.)[8]'].astype(int)
    nfl_scores_metro_subset = nfl_scores_metro_subset.groupby('Metropolitan area').mean().reset_index()
    return nfl_scores_metro_subset
    
def sports_team_performance():
    # YOUR CODE HERE
    nfl = nfl_correlation()
    nba = nba_correlation()
    mlb = mlb_correlation()
    nhl = nhl_correlation()
    # print (nfl.head())
    # print (nhl.head())
    nba_nfl = pd.merge(nba,nfl, on='Metropolitan area')
    print(nba_nfl.head())
    pval_nba_nfl = stats.ttest_rel(nba_nfl['W/L Ratio_x'],nba_nfl['W/L Ratio_y'])[1]
    nba_nhl = pd.merge(nba,nhl, on='Metropolitan area')
    pval_nba_nhl = stats.ttest_rel(nba_nhl['W/L Ratio_x'],nba_nhl['W/L Ratio_y'])[1]
    mlb_nfl = pd.merge(mlb,nfl, on='Metropolitan area')
    pval_mlb_nfl = stats.ttest_rel(mlb_nfl['W/L Ratio_x'],mlb_nfl['W/L Ratio_y'])[1]
    mlb_nhl = pd.merge(mlb,nhl, on='Metropolitan area')
    pval_mlb_nhl = stats.ttest_rel(mlb_nhl['W/L Ratio_x'],mlb_nhl['W/L Ratio_y'])[1]
    mlb_nba = pd.merge(mlb,nba, on='Metropolitan area')
    pval_mlb_nba = stats.ttest_rel(mlb_nba['W/L Ratio_x'],mlb_nba['W/L Ratio_y'])[1]
    nhl_nfl = pd.merge(nhl,nfl, on='Metropolitan area')
    pval_nhl_nfl = stats.ttest_rel(nhl_nfl['W/L Ratio_x'],nhl_nfl['W/L Ratio_y'])[1]
    
    pv = {'NFL': {"NFL": np.nan, 'NBA': pval_nba_nfl, 'NHL': pval_nhl_nfl, 'MLB': pval_mlb_nfl},
       'NBA': {"NFL": pval_nba_nfl, 'NBA': np.nan, 'NHL': pval_nba_nhl, 'MLB': pval_mlb_nba},
       'NHL': {"NFL": pval_nhl_nfl, 'NBA': pval_nba_nhl, 'NHL': np.nan, 'MLB': pval_mlb_nhl},
       'MLB': {"NFL": pval_mlb_nfl, 'NBA': pval_mlb_nba, 'NHL': pval_mlb_nhl, 'MLB': np.nan}
      }
    
    # Note: p_values is a full dataframe, so df.loc["NFL","NBA"] should be the same as df.loc["NBA","NFL"] and
    # df.loc["NFL","NFL"] should return np.nan
    sports = ['NFL', 'NBA', 'NHL', 'MLB']
    p_values = pd.DataFrame({k:np.nan for k in sports}, index=sports)
    
    return p_values
```
